## Pivot

```python
import pandas as pd

var = pd.DataFrame({"days" : [1,2,3,4,5,6],"st_name" : ['Jack', 'Nia', 'Sam','Jack', 'Nia', 'Sam'], "eng" : [10,12,14,15,16,12], "maths" :[17,18,19,13,14,16]})

var
```

```python
# minimum 2 arguments are required
var.pivot(index ='days', columns= 'st_name')
```

```python
var.pivot(index ='days', columns= 'st_name', values='eng')
```

```python
var1 = pd.DataFrame({"days" : [1,2,3,4,5,6],"st_name" : ['Jack', 'Nia', 'Jack','Jack', 'Nia', 'Nia'], "eng" : [10,12,14,15,16,12], "maths" :[17,18,19,13,14,16]})

var1
```

```python
# minimum 2 arguments are required
var1.pivot(index ='days', columns= 'st_name')
```

pivot_table

```python
var2 = pd.DataFrame({"days" : [1,1,1,1,2,2],"st_name" : ['Jack', 'Nia', 'Jack','Jack', 'Nia', 'Nia'], "eng" : [10,12,14,15,16,12], "maths" :[17,18,19,13,14,16]})

var2
```

```python
var2.pivot_table(index='st_name', columns='days', aggfunc = 'mean')
```

```python
var2.pivot_table(index='st_name', columns='days', aggfunc = 'sum')
```

```python
var2.pivot_table(index='st_name', columns='days', aggfunc = 'min')
```
