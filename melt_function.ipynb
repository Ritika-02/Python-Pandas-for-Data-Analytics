melt ()

```python
import pandas as pd

var = pd.DataFrame({"days" : [1,2,3,4,5,6], "eng" : [10,12,14,15,16,12], "maths" :[17,18,19,13,14,16]})

var
```

- The dataframe is horizontal by default to look it vertically we use melt.

```python
# days column is working as id
pd.melt(var, id_vars=['days'])
```

```python
pd.melt(var,id_vars=['days'], var_name='subject', value_name= 'marks')

# var_name : give labels to variable
```
