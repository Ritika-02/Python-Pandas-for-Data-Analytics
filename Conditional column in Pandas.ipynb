### Conditional calculated columns in pandas

#### Scenario 1 : create a new coumn state using city column

#### Scenario 2 : create a new column profit_category using profit column

```python
import mysql.connector
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
conn = mysql.connector.connect(host = 'localhost',password='ritika', user='root', database = 'python')

if conn.is_connected():
    print('Connection established')
```

```python
query1 = 'Select * from orders;'
df_orders = pd.read_sql(query1, conn)

df_orders
```

```python
df_orders.loc[df_orders['city'] == 'Bangalore','state']='Karnataka'
df_orders.loc[df_orders['city'] == 'Chennai','state']='TN'
df_orders
```

```python
## If we have 2 cities from same state
# df_orders.loc[df_orders['city']== isin(['Banglore','Mysore']), 'state']='Karnataka'
```

```python
df_orders.loc[df_orders['profit'] <= 250, 'profit_category']= 'Low profit'
df_orders.loc[(df_orders['profit'] > 250) & (df_orders['profit'] <= 500), 'profit_category']= 'Medium profit'
df_orders.loc[df_orders['profit'] > 500, 'profit_category']= 'High profit'
df_orders
```
