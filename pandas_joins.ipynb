```python
query1 = 'Select * from emp;'
df_emp = pd.read_sql(query1,conn)
```

```python
df_emp
```

```python
query2 = 'Select dept_id as dep_id,dep_name from dept;'
df_dept = pd.read_sql(query2,conn)

df_dept
```

#### Now here we can see table emp has dep_id where as table dept has dept_id now how to join

```python
df_inner_join = pd.merge(left= df_emp, right=df_dept, how= 'inner',on='dep_id')
df_inner_join
```

#### In SQL we have liberty to choose what columns we want to see, but in pandas by default all the columns will come

```python
# what if we have different column names in both of the tables

df_dept.rename(columns={'dep_id':'dept_id'}, inplace= True)
df_dept
```

```python
df_inner_join = pd.merge(left= df_emp, right=df_dept, how= 'inner',left_on='dep_id', right_on='dept_id')
df_inner_join
```

#### By default both dep_id and dept_id are printed

```python
df_left_join = pd.merge(left= df_emp, right=df_dept, how= 'left',left_on='dep_id', right_on='dept_id')
df_left_join
```

```python
df_right_join = pd.merge(left= df_emp, right=df_dept, how= 'right',left_on='dep_id', right_on='dept_id')
df_right_join
```

```python
df_full_outer_join = pd.merge(left= df_emp, right=df_dept, how= 'outer',left_on='dep_id', right_on='dept_id')
df_full_outer_join
```
